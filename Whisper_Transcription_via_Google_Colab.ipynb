```python
# Install Whisper
!pip install -q openai-whisper

# Optional: install FFmpeg (for audio decoding)
!apt-get -y install ffmpeg

# Upload your audio file
from google.colab import files
uploaded = files.upload()

# Transcribe with Whisper
import whisper

model = whisper.load_model("small")  # use "tiny", "base", "small" as you wish
for filename in uploaded.keys():
    print(f"Transcribing {filename}...")
    result = model.transcribe(filename, verbose=True) #result = model.transcribe(filename, language="Indonesian", verbose=True)
    #If your audio is in Indonesian → Whisper gives you Indonesian text automatically.
    print(result["text"])

# Save transcript to a text file
with open("transcript.txt", "w") as f:
    f.write(result["text"])

print("\n✅ Transcription complete! Downloading file...")
files.download("transcript.txt")
```

```python
# === DOWNLOAD RESULT WITH TIMESTAMP ===
from google.colab import files
import json

# Simpan hasil ke JSON (buat simpan semua data Whisper termasuk timestamp)
with open("transcript_with_timestamps.json", "w", encoding="utf-8") as f:
    json.dump(result, f, ensure_ascii=False, indent=2)

# Simpan versi teks biasa tapi ada timestamp-nya
with open("transcript_with_timestamps.txt", "w", encoding="utf-8") as f:
    for seg in result["segments"]:
        start = seg["start"]
        end = seg["end"]
        text = seg["text"].strip()
        f.write(f"[{start:.2f}–{end:.2f}] {text}\n")

# Tampilkan tombol download di Colab
print("✅ File with timestamp ready to download!")
files.download("transcript_with_timestamps.json")
files.download("transcript_with_timestamps.txt")
```
